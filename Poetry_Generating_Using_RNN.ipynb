```python
import numpy as np
import pandas as pd
import string
```

```python
def weights(Mi,Mo):
    return np.random.randn(Mi,Mo)/np.sqrt(Mi+Mo)
```

```python
def remove_punctuation(s):
    translator = str.maketrans('', '', string.punctuation)
    return s.translate(translator)
```

```python
def get_robert_frost():
    word2idx={'Start':0,'End':1}
    current_indx=2
    sentences=[]
    for line in open('RobertFrost.txt'):
            line=line.strip()
            if line:
                line_wp=remove_punctuation(line.lower())
                tokens=line_wp.split()
                sentence=[]
                for t in tokens:
                    if t not in word2idx:
                        word2idx[t] = current_indx
                        current_indx +=1
                    idx=word2idx[t]
                    sentence.append(idx)
                sentences.append(sentence)
    return sentences,word2idx
```
